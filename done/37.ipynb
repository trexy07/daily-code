```python
# this output could be used for simple mountain generation based on depth

# dir(set)

def set_power(s,start=True): # start determins if it is the first pass
    output=[] # have to use a list to put sets in
    if len(s)>1: # no empty sets
        for item in s:
            subset=s-{item} # sets for quick subtraction
            output.append(subset)
    
    # print(output)
    output2=[] # output of second level subsets 
    # (seperation was needed to stop extra loops)
    for subset in output:
        # print("subset",subset)
        output2+=set_power(subset,False) # add the recursive subsets
    
    if start:# on the first pass
        final={tuple(s)}# make a set with the tupled(starting set)
       
        for item in output+output2: # take each of the new outputs
            if item not in final : # remove extras
                item=tuple(item)
                final.add(item)
        
        return final
        # return set(output+output2)
    else:
        return output+output2

set_power({1,2,3,4,5,6,7,8})
```
